```python
x = img.copy()
```

```python
x
```

```python
plt.imshow(np.flipud(x))  #print x flipped up down
```

```python
plt.imshow(np.fliplr(x))    #print x flipped left right
```

```python
plt.imshow(np.ndarray.swapaxes(x,0,1))   #print with x and y axes swapped
```

```python
plt.imshow(x[::-1])  #walk backwards through the y values, flips up down
```

```python
plt.imshow(x[:, ::-1])  #leave the y values alone, then walk backwards through the x values
```
